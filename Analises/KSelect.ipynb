# Analise de resultados de MDC

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import math

```

```python
  
# Opening JSON file
f = open('../test\KSelect\KSelect_results.json')
  
# returns JSON object as 
# a dictionary
data = json.load(f)
df = pd.DataFrame(data)
```

```python
# Opening JSON file
f_insere = open('../test\KSelect\KSelect_results_insere.json')
  
# returns JSON object as 
# a dictionary
data_insere = json.load(f_insere)
df_insere = pd.DataFrame(data_insere)
```

```python
df_insere['Custo Médio de Inserção para Soluciona'] = df_insere.apply(lambda row: row['InsereParaSolucionaGasTotal'] / row['N'], axis=1)
df_insere['Custo Médio de Inserção para Verifica'] = df_insere.apply(lambda row: row['InsereParaVerificaGasTotal'] / row['N'], axis=1)
df_insere['Razão de Custo Soluciona/Verifica'] = df_insere['RazaoSolucionaVerifica']

df_insere[['N','Custo Médio de Inserção para Soluciona', 'Custo Médio de Inserção para Verifica', 'Razão de Custo Soluciona/Verifica']]
```

```python
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(7)

colors = ['#e6194b', '#3cb44b', '#ffe119', '#4363d8', '#f58231','#000000', '#46f0f0', '#f032e6', '#bcf60c', '#fabebe', '#008080', '#e6beff', '#9a6324', '#fffac8', '#800000', '#aaffc3', '#808000', '#ffd8b1', '#000075', '#808080']

for i in range(10):
    plt.plot(df[df.N == 2**(i+1)]['K'].apply(lambda x: math.log(x, 2)), df[df.N == 2**(i+1)]['RazaoSolucionaVerifica'],c=colors[i], label='N=' + str(2**(i+1)),  marker='o')
plt.hlines(y=1, xmin=0, xmax=9, colors='red', linestyles='--', lw=2)

plt.legend()
plt.title("Razão de custo entre Soluciona(X) e Verifica(X,Y,W)", {'fontsize': 14, 'fontweight' : 600})
plt.xlabel("Log K (base 2)")

plt.show()
```

```python
f, ax = plt.subplots()
f.set_figwidth(10)
f.set_figheight(7)



ax.ticklabel_format(useOffset=False, style='plain')

ax.plot(df[df.N == 1024]['K'].apply(lambda x: math.log(x, 2)), df[df.N == 1024]['SolucionaGasInicial'],c='red', label='Soluciona: Gas Inicial ',  marker='o')
ax.plot(df[df.N == 1024]['K'].apply(lambda x: math.log(x, 2)), df[df.N == 1024]['SolucionaGasExecucao'],c='blue', label='Soluciona: Gas Execução ',  marker='o')
ax.plot(df[df.N == 1024]['K'].apply(lambda x: math.log(x, 2)), df[df.N == 1024]['SolucionaGasTotal'],c='black', label='Soluciona: Gas Total ',  marker='o')

ax.plot(df[df.N == 1024]['K'].apply(lambda x: math.log(x, 2)), df[df.N == 1024]['VerificaGasInicial'],c='red', label='Verifica: Gas Inicial ',  marker='o', linestyle='dotted')
ax.plot(df[df.N == 1024]['K'].apply(lambda x: math.log(x, 2)), df[df.N == 1024]['VerificaGasExecucao'],c='blue', label='Verifica: Gas Execução ',  marker='o', linestyle='dotted')
ax.plot(df[df.N == 1024]['K'].apply(lambda x: math.log(x, 2)), df[df.N == 1024]['VerificaGasTotal'],c='black', label='Verifica: Gas Total ',  marker='o', linestyle='dotted')

ax.legend()
plt.title("Custo de Gas para experimento com N=1024", {'fontsize': 14, 'fontweight' : 600})
plt.xlabel("Log K (base 2)")


plt.show()
```
